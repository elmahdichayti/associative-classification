# Tests :

### Titanic Data set

```julia
dataSet = "titanic"
dataFolder = "data/"
resultsFolder = "res/"

# Create the features tables (or load them if they already exist)
# Note: each line corresponds to an individual, the 1st column of each table contain the class
train, test = createFeatures(dataFolder, dataSet)
```

```julia
rules = createRules(dataSet, resultsFolder, train)
```

```julia
orderedRules = sortRules(dataSet, resultsFolder, train, rules)
```

```julia
recall = getPrecision(orderedRules, test)
accuracy = getPrecision(orderedRules, train)

println("Recall: ", recall)
println("Accuracy: ", accuracy)
```

### MultiJuMP :

```julia
rules_multijump = createRules_withMultiJuMP(dataSet, resultsFolder, train)
```

```julia
orderedRules_MultiJuMP = sortRules_MultiJuMP(dataSet, resultsFolder, train, rules_multijump)
```

```julia
recall = getPrecision(orderedRules_MultiJuMP, test)
accuracy = getPrecision(orderedRules_MultiJuMP, train)

println("Recall: ", recall)
println("Accuracy: ", accuracy)
```

### Haberman Data Set

```julia
dataSet = "haberman"
dataFolder = "data/"
resultsFolder = "res/"

# Create the features tables (or load them if they already exist)
# Note: each line corresponds to an individual, the 1st column of each table contain the class
train_haber, test_haber = createFeatures(dataFolder, dataSet)
```

```julia
# Create the rules (or load them if they already exist)
# Note: each line corresponds to a rule, the first column corresponds to the class
rules_haber = createRules(dataSet, resultsFolder, train_haber)
```

```julia
orderedRules_haber = sortRules(dataSet, resultsFolder, train_haber, rules_haber)
```

```julia
recall_haber = getPrecision(orderedRules_haber, test_haber)
accuracy_haber = getPrecision(orderedRules_haber, train_haber)

println("Haberman's Recall: ", recall_haber)
println("Haberman's Accuracy: ", accuracy_haber)
```
